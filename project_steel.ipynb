```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None

df_orders = pd.read_excel('C:/Users/vp.mishnev/Desktop/аналитика/data2.xlsx',usecols = ['Дата-время начала', 'Дата-время окончания', 'Марка'], skiprows = range(2), nrows = 100 )
```

```python
df_orders['Дата-время окончания'] = pd.to_datetime(df_orders['Дата-время окончания'])
df_orders['Темп работы'] = 0
df_orders['Темп работы'] = (df_orders['Дата-время окончания'] - df_orders['Дата-время окончания'].shift(1))
df_orders['Темп работы'] = df_orders['Темп работы'].dt.total_seconds()
df_orders['Темп работы'] = df_orders['Темп работы'].fillna(0)
```

```python

df_orders['Номер массива'] = 0.0
ls = [0]
counter = 0
for i in range(1, len(df_orders)):
    num = df_orders['Темп работы'].shift(1).iloc[i]
    if num>10:
        counter = counter+1
    ls.append(counter)
df_orders['Номер массива'] = ls
```

```python
#занесём среднее время в начу таблицу
df_orders['Время полистно'] = df_orders.groupby('Номер массива')['Темп работы'].transform('mean')
```

```python
df_steel = pd.read_excel('C:/Users/vp.mishnev/Desktop/аналитика/Производительность расчёт 23-24 6 мес — копия — копия.xlsx', sheet_name = 2)
```

```python
df_steel.rename(columns = {'Марки сталей': 'Марка'}, inplace = True)
df_orders = pd.merge(df_orders, df_steel, on = 'Марка', how = 'left')
```
